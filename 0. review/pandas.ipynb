### DataFrame 생성

```python
dic1 = {'이름' : ['기범', '민호', '태민', '진기'],
        '년도' : [1991, 1991, 1993, 1989],
        '동물' : ['햄스터', '다람쥐', '병아리', '토끼']
       }

# 딕셔너리를 DataFrame으로 변환
df = pd.DataFrame(dic1)
display(df)


# 새로운 컬럼명을 추가
df = pd.DataFrame(dic1, columns=["이름", "년도", "동물", "나이"])
display(df)


# 인덱스를 새로운 값으로 할당. 
df = pd.DataFrame(dic1, index=['one','two','three','four'])
display(df)
```

### DataFrame과 list, dictionary, numpy ndarray 상호변환

* list, dict, ndarray ====> dataframe

```python
import numpy as np
```

```python
col_name1=['col1']
list1 = [1,2,3]
array1 = np.array(list1)

print('array1 shape:', array1.shape)

# list를 df로
df_list1 = pd.DataFrame(list1, columns = col_name1)
# ndarray를 df로
df_array1 = pd.DataFrame(array1, columns = col_name1)

print(df_list1)
print(df_array1)
```

```python
# 3개의 칼럼명이 필요한 경우
col_name2 = ['col1', 'col2','col3']

# 2행 by 3열 형태의 list와 ndarray를 df로
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)

print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns = col_name2)
df_array2 = pd.DataFrame(array2, columns = col_name2)

print(df_list2)
print(df_array2)
```

```python
# key는 column명으로 매핑, value는 list형 또는 ndarray
dict = {'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print(df_dict)
```

* dataframe =====> ndarray, list, dict

```python
# df ==> ndarray 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3))
print(array3)
```

```python
# df ==> list 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# df ==> dict 변환
dict3 = df_dict.to_dict('list')
print('df_dict.to_dict() 타입:', type(dict3))
print(dict3)

df_dict.to_dict()
```

### DataFrame의 칼럼 데이터 세트 생성 및 수정

```python
titanic_df['Age_0'] = 0
titanic_df['Age_by_10'] = titanic_df['Age']*10
titanic_df['Family_No'] = titanic_df['SibSp'] + titanic_df['Parch']+1
titanic_df.head(3)
```

### DataFrame 데이터 삭제
**axis**에 따른 삭제
* row 삭제 : axis = 0, col 삭제 : axis = 1
* 원본 df는 유지하고 drop된 df를 새롭게 객체 변수로 받고 싶다면 inplace=False로 설정
* 원본 df에 직접 drop된 결과를 적용할 경우에는 inplace=True를 적용
* 아니면 원본 df와 동일한 객체로 저장

```python
titanic_drop_df = titanic_df.drop('Age_0', axis=1 )
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
```

```python
titanic_df.drop('Age_0', axis=1, inplace=True)
titanic_df.head(3)
```

```python
# 여러 열을 삭제할 때, column명을 적는다. 
titanic_df.drop(['Age_by_10', 'Family_No'], axis=1, inplace=True)
titanic_df.head(3)
```

```python
# 여러 행을 삭제할 때, index를 적는다
titanic_df.drop([0,1,2], axis=0, inplace=True)
titanic_df.head(3)
```

### DataFrame 인덱싱 및 필터링

* df의 [ ] 연산자

```python
# DataFrame객체에서 []연산자내에 한개의 컬럼만 입력하면 Series 객체를 반환  
series = titanic_df['Name']
print(series.head(3))
print("## type:",type(series), 'shape:', series.shape)

# DataFrame객체에서 []연산자내에 여러개의 컬럼을 리스트로 입력하면 그 컬럼들로 구성된 DataFrame 반환  
filtered_df = titanic_df[['Name', 'Age']]
display(filtered_df.head(3))
print("## type:", type(filtered_df), 'shape:', filtered_df.shape)

# DataFrame객체에서 []연산자내에 한개의 컬럼을 리스트로 입력하면 한개의 컬럼으로 구성된 DataFrame 반환 
one_col_df = titanic_df[['Name']]
display(one_col_df.head(3))
print("## type:", type(one_col_df), 'shape:', one_col_df.shape)
```
