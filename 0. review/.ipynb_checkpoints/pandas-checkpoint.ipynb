### DataFrame과 list, dictionary, numpy ndarray 상호변환

* list, dict, ndarray ====> dataframe

```python
import numpy as np
```

```python
col_name1=['col1']
list1 = [1,2,3]
array1 = np.array(list1)

print('array1 shape:', array1.shape)

# list를 df로
df_list1 = pd.DataFrame(list1, columns = col_name1)
# ndarray를 df로
df_array1 = pd.DataFrame(array1, columns = col_name1)

print(df_list1)
print(df_array1)
```

```python
# 3개의 칼럼명이 필요한 경우
col_name2 = ['col1', 'col2','col3']

# 2행 by 3열 형태의 list와 ndarray를 df로
list2 = [[1,2,3],
         [11,12,13]]
array2 = np.array(list2)

print('array2 shape:', array2.shape)

df_list2 = pd.DataFrame(list2, columns = col_name2)
df_array2 = pd.DataFrame(array2, columns = col_name2)

print(df_list2)
print(df_array2)
```

```python
# key는 column명으로 매핑, value는 list형 또는 ndarray
dict = {'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print(df_dict)
```

* dataframe =====> ndarray, list, dict

```python
# df ==> ndarray 변환
array3 = df_dict.values
print('df_dict.values 타입:', type(array3))
print(array3)
```

```python
# df ==> list 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

# df ==> dict 변환
dict3 = df_dict.to_dict('list')
print('df_dict.to_dict() 타입:', type(dict3))
print(dict3)

df_dict.to_dict()
```

### DataFrame 데이터 삭제
**axis**에 따른 삭제
* row 삭제 : axis = 0, col 삭제 : axis = 1
* 원본 df는 유지하고 drop된 df를 새롭게 객체 변수로 받고 싶다면 inplace=False로 설정
* 원본 df에 직접 drop된 결과를 적용할 경우에는 inplace=True를 적용
* 아니면 원본 df와 동일한 객체로 저장

```python
titanic_drop_df = titanic_df.drop('Age_0', axis=1 )
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
```

```python
titanic_df.drop('Age_0', axis=1, inplace=True)
titanic_df.head(3)
```

```python
# 여러 열을 삭제할 때, column명을 적는다. 
titanic_df.drop(['Age_by_10', 'Family_No'], axis=1, inplace=True)
titanic_df.head(3)
```

#### groupby( ) 이용하기

* groupby( ) 내의 인자로 by = groupby하고자 하는 칼럼명 입력

dataframe에 groupby()를 적용하면 DataFrameGroupBy 객체를 반환함

```python
titanic_groupby[['Age', 'Fare']].mean()
```

```python
titanic_groupby = titanic_df.groupby('Pclass').count()
titanic_groupby
```

```python
titanic_groupby = titanic_df.groupby('Pclass')[['PassengerId','Survived']].count()
titanic_groupby
```

```python
# 여러 종류의 agg function을 적용하는 것도 가능, 예를 들어 Pclass별 Age의 min값과 max값을 동시에 관찰하고 싶다면?
titanic_groupby = titanic_df.groupby('Pclass')['Age'].agg([max, min])
titanic_groupby
```

```python
# 아니면 컬럼별로 다른 agg function을 적용하고 싶은 경우에는?
aggformat = {'Age':'max', 'SibSp':'sum', 'Fare':'mean'}

titanic_df.groupby('Pclass').agg(aggformat)
# Pclass별로 ~ age에 대해서는 최댓값, sibsp에 대해서는 합계, fare에 대해서는 평균값을 반환
```

### apply lambda 식으로 데이터 가공

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
# 조건을 주는 방식

titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <=15 else 'Adult')
```

```python
titanic_df[['Age', 'Child_Adult']].head(10)
```

```python
titanic_df['category'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['category'].value_counts()
```

```python
def category(age):
    if age <= 5:
        return 'Baby'
    elif age <=12:
        return 'Child'
    elif age <=18:
        return 'Teenager'
    elif age<=35:
        return 'Adolescent'
    elif age<=60:
        return 'Adult'
    else:
        return 'Elderly'
    
titanic_df['category'] = titanic_df['Age'].apply(lambda x : category(x))
titanic_df[['category', 'Age']].head(10)
```
