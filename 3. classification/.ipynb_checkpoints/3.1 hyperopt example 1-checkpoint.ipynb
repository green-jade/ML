```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import warnings

warnings.filterwarnings('ignore')
df = pd.read_csv('santander.csv', encoding='latin-1')
df.head(3)
```

```python
df.TARGET.value_counts()
```

```python
df.describe()
```

```python
df.var3.value_counts()
```

```python
df.var3.replace(-999999, 2, inplace=True)
df.drop(columns=['ID'], axis=1, inplace=True)

X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
y_test.value_counts()/y_test.count()
```

```python
y_train.value_counts()/y_train.count()
```

```python
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=0)
```

### 기억하자 !!

**validation** 검증 데이터 세트는 **조기중단**의 검증 데이터로 이용하기 위해 만드는 것.

```python
# 먼저 XGBoost 학습 모델 생성
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score

# 모델 생성
xgb_clf = XGBClassifier(n_estimators=500, learning_rate=0.05, random_state=100)
# 학습 수행
xgb_clf.fit(X_tr, y_tr, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_val,y_val)])
```

```python
# 평가
xgb_clf.predict_proba(X_test)[:,1]
```
