## (2). 하이퍼파라미터 튜닝 진행

#### XGBoost에 이용되는 하이퍼파라미터 종류!

* learning_rate : (scikit-learn wrapper의 경우) 디폴트 0.1
* n_estimators : 반복 수행 횟수
* min_child_weight : 트리에서 추가적으로 분할할지 결정하는 weight의 총합, 디폴트는 1, 클수록 분할 자제
* max_depth : 트리의 최대 깊이
* colsample_bytree : 트리 생성에 쓰이는 피처 종류를 제한해서 과적합 방지

```python
from sklearn.model_selection import KFold
```

```python
from hyperopt import hp

# 1. 하이퍼파라미터 검색 공간 만들기

search_space = {'max_depth':hp.quniform('maxdepth',5,15,1),
                'min_child_weight':hp.quniform('min_child_weight',1,6,1),
                'colsample_bytree':hp.uniform('colsample_bytree',0.5,0.95),
                'learning_rate': hp.uniform('learning_rate',0.01,0.2)}

# 2. 목적함수 정의하기

def objective_func(search_space):
    xgb_clf = XGBClassifier(n_estimators=100, max_depth=int(search_space['max_depth']),
                            min_child_weight=int(search_space['min_child_weight']),
                            colsample_bytree=search_space['colsample_bytree'],
                            learning_rate=search_space['learning_rate']
                           )
    # 3개 k-fold 방식으로 평가된 roc_auc 지표를 담는 list
    roc_auc_list= []
    
    # 3개 k-fold방식 적용 
    kf = KFold(n_splits=3)
    # X_train을 다시 학습과 검증용 데이터로 분리
    for tr_index, val_index in kf.split(X_train):
        # kf.split(X_train)으로 추출된 학습과 검증 index값으로 학습과 검증 데이터 세트 분리 
        X_tr, y_tr = X_train.iloc[tr_index], y_train.iloc[tr_index]
        X_val, y_val = X_train.iloc[val_index], y_train.iloc[val_index]
        # early stopping은 30회로 설정하고 추출된 학습과 검증 데이터로 XGBClassifier 학습 수행. 
        xgb_clf.fit(X_tr, y_tr, early_stopping_rounds=30, eval_metric='auc',
                   eval_set=[(X_tr, y_tr), (X_val, y_val)])
    
        # 1로 예측한 확률값 추출후 roc auc 계산하고 평균 roc auc 계산을 위해 list에 결과값 담음. 
        score = roc_auc_score(y_val, xgb_clf.predict_proba(X_val)[:, 1])
        roc_auc_list.append(score)
        
    # 3개 k-fold로 계산된 roc_auc값의 평균값을 반환하되, 
    # HyperOpt는 목적함수의 최소값을 위한 입력값을 찾으므로 -1을 곱한 뒤 반환. 
    return -1 * np.mean(roc_auc_list)
```

```python
# 3. 반환값이 최소가 되는 하이퍼파라미터 찾기

from hyperopt import fmin, tpe, Trials

trials=Trials()

best=fmin(fn=objective_func,
          space=search_space,
          algo=tpe.suggest,
          max_evals=10, #최대 반복 횟수
          trials=trials)

print(best)
```

```python
# 4. 최적 하이퍼파라미터를 바탕으로 모델 재구성, test data로 AUC 측정

xgb_clf = XGBClassifier(n_estimators=50,
                            learning_rate=best['learning_rate'],
                            max_depth=int(best['maxdepth']),
                            min_child_weight=int(best['min_child_weight']),
                            colsample_bytree=best['colsample_bytree']
                            )

xgb_clf.fit(X_tr, y_tr, early_stopping_rounds=100,
            eval_metric='auc',eval_set=[(X_val,y_val)])

xgb_roc_score=roc_auc_score(y_test,xgb_clf.predict_proba(X_test)[:,1])
print(xgb_roc_score)
```

*evaluation 횟수를 튜닝 전과 동일하게 지정하지 않아 오히려 auc가 작아졌지만 원래대로라면 성능의 개선이 이루어짐*

### (2) 하이퍼파라미터 튜닝 진행

```python
search_space_lgbm = {'num_leaves':hp.quniform('num_leaves', 32, 64, 1),
                     'max_depth' :hp.quniform('max_depth' , 100,160,1),
                     'min_child_samples':hp.quniform('min_child_samples', 60, 100, 1),
                     'subsample':hp.uniform('subsample', 0.7, 1),
                     'learning_rate':hp.uniform('learning_rate',0.01, 0.2)}

def objective_fun(search_space_lgbm):
    lgbm_clf = LGBMClassifier(n_estimators=100,
                              num_leaves=int(search_space_lgbm['num_leaves']),
                              max_depth=int(search_space_lgbm['max_depth']),
                              min_child_samples=int(search_space_lgbm['min_child_samples']),
                              subsample=search_space_lgbm['subsample'],
                              learning_rate=search_space_lgbm['learning_rate'])
    auc_list = []
    
    kfolds=KFold(n_splits=3)
    
    for tr_index, val_index in kfolds.split(X_train):
        X_tr, y_tr = X_train.iloc[tr_index], y_train.iloc[tr_index]
        X_val, y_val = X_train.iloc[val_index], y_train.iloc[val_index]
        
        lgbm_clf.fit(X_tr, y_tr, early_stopping_rounds=100, eval_metric='auc', eval_set=[(X_val,y_val)])
        score = roc_auc_score(y_val, lgbm_clf.predict_proba(X_val)[:, 1])
        auc_list.append(score)
        
    return -1 * np.mean(auc_list)
```

```python
trials=Trials()

best = fmin(fn = objective_fun, 
            space = search_space_lgbm,
            max_evals=50,
            algo=tpe.suggest,
            trials=trials,
            rstate=np.random.default_rng(seed=30))

print(best)
```
